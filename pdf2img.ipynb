```python
import os
import glob
from shutil import copyfile
from pdf2image import convert_from_path
```

```python
cnt = 0
for pdf_file in glob.glob('/Users/sindhukumar/taiger_iterations/classification_osl/data/valid/*'):
    if not pdf_file.lower().endswith('.pdf'):
        continue
    cnt+=1
#     pdf_path = os.path.join(pdfs_dir, pdf_file)
    try:
#         print(cnt, ": ",pdf_file)
        imgs = convert_from_path(pdf_file)

        for idx, img in enumerate(imgs):
            out_path = pdf_file.replace('.pdf','')+'_'+str(idx)+'.png'
            img.save(out_path, dpi=(400,400))
        os.remove(pdf_file)
    except:
        os.remove(pdf_file)
```
