```python
import pandas
import csv
import numpy as np
import os
import cv2
import matplotlib.pyplot as plt
```

```python
gts = pandas.read_csv('../passport/data/pp_big.csv')
gts[:5]
```

```python
gts_dict = {}
for index, row in gts.iterrows():
    filename = row[0]
    bbox = eval(row[5])
    if len(bbox.keys()) == 0:
        continue
    box = [bbox['x'], bbox['y'], bbox['width'], bbox['height']]
    if filename in gts_dict:
        gts_dict[filename].append(box)
    else:
        gts_dict[filename] = [box]
gts_dict
```

```python
for file, boxes in gts_dict.items():
    img = cv2.imread('../passport/data/big/'+file)
    imh, imw = img.shape[:2]
    with open('../passport/data/labels_v2/'+os.path.splitext(file)[0]+'.txt', 'a') as f:
        for box in boxes:
            x1 = float(box[0])
            y1 = float(box[1])
            wd = float(box[2])
            ht = float(box[3])
            x = (x1+(wd/2.0))/imw
            y = (y1+(ht/2.0))/imh
            w = wd/imw
            h = ht/imh
            text = '0'+' '+str(x)+' '+str(y)+' '+str(w)+' '+str(h)+'\n'
            f.write(text)  
    cv2.imwrite('../passport/data/imgs_v2/' + os.path.splitext(file)[0]+'.jpg', img)
```
